```python
#sort multiple columns

sorted_data_multiple=df.sort_values(by=['sepal_length','sepal_width'],ascending=[False,True])
print("\n Sorted data by sepal length (descending) Sorted data by sepal length (ascending)")
print(sorted_data_multiple.head())
```

```python
#filtering the dataset
#sekect rows where petal length > 1.5
filtered_data=df[df['petal_length']>1.5]
print('\n Filtered data (petal_length > 1.5):')
filtered_data.head()
```

```python
#Grouping by species and calculating means 
grouped_data=df.groupby('species').mean()
print("\n Mean values grouped by species:")
print(grouped_data)
```

```python
df["sepal_area"]=df["sepal_length"]*df["sepal_width"]
print("\n Dataset with sepal_area column added:")
df.head()
```

```python
# example: function to classify sepal length into categories
def classify_sepal_length(length):
    if length < 5.5:
        return 'short'
    elif length <6.5:
        return 'medium'
    else:
        return 'long'
        
#apply function to create a new column 'sepal_length_category'
df['sepal_length_category']=df['sepal_length'].apply(classify_sepal_length)
print("\n Dataset with sepal_length_category column added:")
print(df.head())
```

```python
pivot_table=pd.pivot_table(df,values = 'petal_width' ,index='species',columns='sepal_length_category',aggfunc='mean')
print("\n Pivot table of mean petal_width and sepal_length_category")
pivot_table
```

```python
# encode species column with one hot encoding: One hot encoding is not necessary for tree based models

encoded_data=pd.get_dummies(df,columns=['species'])
print("\n Datasets with species column encoded:")
print(encoded_data.head())
```

```python
concat_data=pd.concat([df,df.copy()],ignore_index=True)
print("Concatenated Data:")
concat_data.head()
```

```python
#reshape data using melt
melted_data=pd.melt(df,id_vars=['species'],
                    value_vars=['sepal_length','sepal_width','petal_length','petal_width'])
print("\nMelted Dataset:")
print(melted_data.head(80))
                                
```

```python
#check the numnber of duplicates
duplicates=data.duplicated().sum()
print(f"Number of duplicate rows :{duplicates}")
```

```python
data.isnull().sum()
```

```python

missing_columns={}
for col in data.columns:
    missing_count=data[col].isnull().sum()
    if missing_count > 0:
        missing_percentage=(missing_count/len(data))*100
        missing_columns[col]=round(missing_percentage, 2)

print("\nMissing coulumns with percentage is ")

for col,percentage in missing_columns.items():
    print(f'{col}: {percentage}%')
```
